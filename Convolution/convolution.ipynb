# pad image function

```python
def pad_image(img, struc_size):
    # pad image with (struc size / 2) zeros 
    pad_size = int(struc_size / 2)
    print(pad_size)
    height, width = img.shape[:2]
    padded_img = np.zeros(shape=(height+pad_size, width+pad_size))
    for i in range(pad_size, height):
        for j in range(pad_size, width):
            padded_img[i,j] = img[i-1,j-1]
    plt.imshow(padded_img)
    return padded_img
```
